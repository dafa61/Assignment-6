```python
from sklearn import tree
import numpy as np
import pandas as pd
```

```python
datasets = pd.read_csv("INTELLIGENT IRRIGATION SYSTEM.csv")
```

```python
datasets.head()
```

```python
# Merubah kelas/kolom dari String ke Unique-Integer
datasets['crop'] = pd.factorize(datasets.crop)[0]
```

```python
# # # Menghapus kolom ID
# datasets = datasets.drop(labels="Pregnancies", axis=1)
```

```python
#Merubah dataframe ke array numpy
datasets = datasets.to_numpy() 
```

```python
print(datasets)
```

```python
#Membagi datasets => 400 baris data untuk training dan 100 baris data untuk testing
dataTraining = np.concatenate((datasets[0:80, :], datasets[100:180, :]), axis=0)
dataTesting = np.concatenate((datasets[80:100, :], datasets[180:200, :]), axis=0)
```

```python
print(dataTraining)
len(dataTraining)
```

```python
print(dataTesting)
len(dataTesting)
```

```python
#Memecah dataset ke input label
inputTraining = dataTraining[:,0:3]
inputTesting = dataTesting[:,0:3]
labelTraining = dataTraining[:,3]
labelTesting = dataTesting[:,3]
print(labelTraining)
len(labelTraining)
```

```python
#Mendefinisikan decision tree classifier
model = tree.DecisionTreeClassifier()
```

```python
#Mentraining model
model =  model.fit(inputTraining, labelTraining)
```

```python
#Memprediksi input data testing
hasilPrediksi = model.predict(inputTesting)
print("Label Sebenarnya : ", labelTesting)
print("Hasil Prediksi : ", hasilPrediksi)
```

```python
#Menghitung akurasi
prediksiBenar = (hasilPrediksi == labelTesting).sum()
prediksiSalah = (hasilPrediksi != labelTesting).sum()
print("Prediksi Benar : ", prediksiBenar, " Data")
print("Prediksi Salah : ", prediksiSalah, " Data")
print("Akurasi : ", prediksiBenar/(prediksiBenar+prediksiSalah) * 100, "%")
```
